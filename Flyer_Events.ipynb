# **1. Computation of unique users**

```python
unique_users = data["user_id"].nunique()
print("Number of unique users:", unique_users)
```

# **2. Computation of average time spent on flyers per user**

```python
flyer_events = data[data["event"] == "flyer_open"].sort_values(["user_id", "timestamp"])
flyer_events.head()
```

```python
flyer_events["next_timestamp"] = flyer_events.groupby("user_id")["timestamp"].shift(-1)
flyer_events.head()
```

```python
flyer_events["time_spent"] = (flyer_events["next_timestamp"] - flyer_events["timestamp"]).dt.total_seconds()  # Unit of time spent = Seconds
flyer_events.head()
```

```python
flyer_events = flyer_events.dropna(subset=["time_spent"]) # Remove entries where next_timestamp is NaN (last event for a user)

avg_time_per_user = flyer_events.groupby("user_id")["time_spent"].sum().mean()
print("Average time spent on flyers per user:", avg_time_per_user)
```

# **3. Computation of average time spent on flyer_id = 2020004**

```python
flyer_events['flyer_id']=flyer_events["flyer_id"].astype(int)   # change data type from object to int
flyer_2020004 = flyer_events[flyer_events["flyer_id"] == 2020004]
avg_time_flyer_2020004 = flyer_2020004["time_spent"].mean()
print("Average time spent on flyer_id = 2020004:", avg_time_flyer_2020004)
```

# **Computation of above three tasks through a function**

```python
def flyer_data_tasks(data):
  import pandas as pd
  from datetime import datetime
  data["timestamp"] = pd.to_datetime(data["timestamp"])
  unique_users = data["user_id"].nunique()
  flyer_events = data[data["event"] == "flyer_open"].sort_values(["user_id", "timestamp"])
  flyer_events["next_timestamp"] = flyer_events.groupby("user_id")["timestamp"].shift(-1)
  flyer_events["time_spent"] = (flyer_events["next_timestamp"] - flyer_events["timestamp"]).dt.total_seconds()
  flyer_events = flyer_events.dropna(subset=["time_spent"])
  avg_time_per_user = flyer_events.groupby("user_id")["time_spent"].sum().mean()
  flyer_events['flyer_id']=flyer_events["flyer_id"].astype(int)
  flyer_2020004 = flyer_events[flyer_events["flyer_id"] == 2020004]
  avg_time_flyer_2020004 = flyer_2020004["time_spent"].mean()
  return unique_users, avg_time_per_user, avg_time_flyer_2020004
```

```python
flyer_data_tasks(data)
```

# **Explanation of algorithm scales for:  1 Million Events (~10 MB of data)**

**Memory Usage:**

The dataset is relatively small and can be processed entirely in memory using pandas. Standard machines (e.g., 8-16 GB RAM) can easily handle this volume of data.


**Performance:**

Operations like groupby, sort, and shift are optimized in pandas for datasets of this size. Runtime for the entire computation would typically be in the range of a few seconds to a minute, depending on the machine and implementation efficiency.

**Environment:**

No special distributed computing tools are needed. The entire workflow can be executed on a single machine.

# **Explanation of algorithm scales for:  1 Trillion Events (~1 TB of data)**

**Memory Usage:**

This scale far exceeds the memory capacity of standard machines. Processing must be done using distributed computing platforms like Apache Spark or Dask. The data would likely be stored in a distributed filesystem (e.g., HDFS or S3).

**Processing Framework:**

Frameworks like Spark are designed for distributed processing and can handle groupby and sort operations in parallel across multiple nodes.
Data partitioning by user_id or flyer_id is essential to distribute the workload efficiently.

**Optimization Techniques:**

Incremental Computation: Process data in batches to reduce memory usage.
Aggregation: Pre-aggregate data by user or flyer within smaller batches before merging results globally.
Efficient Storage: Use columnar data formats (e.g., Parquet or ORC) to speed up read/write operations and minimize storage overhead.

**Performance Expectations:**

Distributed systems handle this scale efficiently, but runtime can vary from minutes to hours depending on hardware, cluster size, and optimization level.

**Scalability Strategy:**

Add more compute nodes to the cluster to handle increases in data size.
Use caching and partition pruning for frequently accessed subsets of data.
